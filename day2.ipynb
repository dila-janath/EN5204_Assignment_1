# intensity transformation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im_orig = cv.imread('flowers.jpeg', cv.IMREAD_GRAYSCALE)
t = 255 - np.arange(0,256,1).astype(np.uint8)
assert t.shape[0] == 256


im2 = t[im_orig]

plt.plot(t)

cv.namedWindow('Display', cv.WINDOW_NORMAL)
cv.imshow('Display',im_orig)
cv.waitKey()

cv.imshow('Display',im2)
cv.waitKey()
cv.destroyAllWindows()


```

# histograms

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im_orig = cv.imread('butter.jpeg', cv.IMREAD_COLOR)
#im_orig[:,:,2:3] = 0
hist_b = cv.calcHist([im_orig],[0], None, [256], [0,256])
hist_g = cv.calcHist([im_orig],[1], None, [256], [0,256])
hist_r = cv.calcHist([im_orig],[2], None, [256], [0,256])

plt.plot(np.log(hist_b), 'b')
plt.plot(np.log(hist_g), 'g')
plt.plot(np.log(hist_r), 'r')

#plt.plot(hist_g)
#plt.plot(hist_r)

plt.xlim([0,256])
plt.show()


```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im_orig = cv.imread('equlhist.png', cv.IMREAD_GRAYSCALE)
hist = cv.calcHist([im_orig], [0], None, [256], [0,256])
plt.plot(hist)

equ = cv.equalizeHist(im_orig)

plt.xlim([0,256])
plt.show()

hist2 = cv.calcHist([equ], [0], None, [256], [0,256])
plt.plot(hist2)

plt.xlim([0,256])
plt.show()




fig,ax = plt.subplots(1,2, figsize=(8,16))
ax[0].imshow(im_orig, cmap='gray', vmin=0, vmax=255)
ax[0].xaxis.tick_top()
ax[1].imshow(equ, cmap='gray', vmin=0, vmax=255)
ax[1].xaxis.tick_top()
plt.show
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('butter.jpeg', cv.IMREAD_COLOR)
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)
gamma = 2

t1 = np.array([(i/255)**gamma*255 for i in range (0,256)]).astype(np.uint8)
im2 = t1[im]

gamma = 0.5

t2 = np.array([(i/255)**gamma*255 for i in range (0,256)]).astype(np.uint8)
im3 = t2[im]

fig,ax = plt.subplots(1,3, figsize=(8,16))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].xaxis.tick_top()
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
ax[1].xaxis.tick_top()
ax[2].imshow(im3, cmap='gray', vmin=0, vmax=255)
ax[2].xaxis.tick_top()

plt.show
```
